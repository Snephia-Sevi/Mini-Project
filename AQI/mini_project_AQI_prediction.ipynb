```python
#for numerical computing
import numpy as np
#for dataframes
import pandas as pd
#Ignore Warnings
import warnings
warnings.filterwarnings("ignore")
#to split train and test set
from sklearn.model_selection import train_test_split
#data preprocessing
from sklearn.preprocessing import StandardScaler, LabelEncoder
from collections import Counter

import sklearn.metrics as sm
```

```python
df=pd.read_csv('city_day.csv')
print(df.shape)
print(df.columns)
#df['Date'] = pd.to_datetime(df['Date'])
#df.rename(columns = {'AQI_Bucket':'Air_quality'}, inplace = True)
df.info()





```

```python

df.describe()
```

```python
print(df.head())
print(df.describe())
print(df)
```

```python
df=df.drop_duplicates()
```

```python
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna()
```

```python
print(df.isnull().sum())
df.drop(["City","Date","AQI_Bucket"],axis=1,inplace=True)

df.describe()
print(df)
```

```python
print(df.corr())
cor = df.corr()
cor.style.background_gradient(cmap='coolwarm')
```

```python
y = df["AQI"]
x = df[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2','O3', 'Benzene', 'Toluene', 'Xylene']]
#Split X and y into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python

print(X_test)
```

```python

print('Classes and number of values in trainset',Counter(y_train))
```

```python
#Print number of observations in X_train,X_test,y_train and y_test
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
model1=RandomForestRegressor()
from sklearn.linear_model import LinearRegression
model2=LinearRegression()

model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
```

```python
#Predict Test set results
y_pred1=model1.predict(X_test)
y_pred2=model2.predict(X_test)
```

```python
df1=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred1})
print (df1)
```

```python
df2=pd.DataFrame({'Actual':y_test , 'Predicted':y_pred2})
print(df2)
```

```python
y_pred1= np.array(y_pred1).reshape(-1,1)
y_pred2= np.array(y_pred2).reshape(-1,1)

accuracy= sm.r2_score(y_test,y_pred1)
print("Accuracy of RandomForestRegressor is {:.2f} % ".format(accuracy*100))
```

```python
accuracy=sm.r2_score(y_test,y_pred2)
print("Accuracy of Linear Regression is {:.2f} % ".format(accuracy*100))
```
